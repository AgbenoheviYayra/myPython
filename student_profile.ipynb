```python
name = input("What is your Name?")
age = int(input("What is your Age?"))

gpa = float(input ("What is your GPA?"))

if gpa > 4.0:
    print ("Invalid gpa")
elif gpa >= 3.6:
    deans_list = "Yes"
else:
    deans_list = "No"

courses_input = input("Enter your courses (comma-separated): ")
courses = [course.strip() for course in courses_input.split(",")]


student_profile = {}
student_profile ["name"] = name
student_profile ["age"] = age
student_profile ["gpa"] = gpa
student_profile ["num_courses"] = courses
student_profile ["Deans_List"] = deans_list

print("========STUDENT PROFILE========")
print(f"Name: {name}")
if age >= 18:
    print (f"Age: {age} (Adult)")
else:
    print (f"Age: {age} (Minor)")
print(f"GPA: {gpa}")
print(f"Deans List: {deans_list}")

print(f"Courses: {courses}")
print("===============================")
```
